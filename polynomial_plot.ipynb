```python
import numpy as np
import matplotlib.pyplot as plt

def polynomial_equation(coefficients):
    # Create the polynomial
    poly = np.poly1d(coefficients)
    
    # Get the equation as a string
    equation_str = ""
    for power, coeff in enumerate(reversed(poly.coefficients)):
        if coeff == 0:
            continue  # Skip zero coefficients
        elif power == 0:
            equation_str += f"{coeff:.2f}"
        elif power == 1:
            equation_str += f" + {coeff:.2f}x"
        else:
            equation_str += f" + {coeff:.2f}x^{power}"
    
    # Return the equation
    return equation_str.strip(" +")

def plot_polynomial(coefficients, x_range=(-10, 10)):
    # Generate x values
    x = np.linspace(x_range[0], x_range[1], 400)
    # Create the polynomial
    poly = np.poly1d(coefficients)
    # Evaluate the polynomial
    y = poly(x)
    
    # Plot the polynomial
    plt.plot(x, y, label=f'Equation: {polynomial_equation(coefficients)}')
    plt.title(f'Polynomial Plot')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend(loc='upper left')
    plt.grid(True)
    plt.show()

def polynomial_equation_latex(coefficients):
    # Create the polynomial
    poly = np.poly1d(coefficients)
    
    # Build the LaTeX string
    equation_str = "$y = "  # Start with LaTeX math mode
    for power, coeff in enumerate(reversed(poly.coefficients)):
        if coeff == 0:
            continue  # Skip zero coefficients
        elif power == 0:
            equation_str += f"{coeff:.2f}"
        elif power == 1:
            equation_str += f" + {coeff:.2f}x"
        else:
            equation_str += f" + {coeff:.2f}x^{power}"
    
    equation_str += "$"  # End the LaTeX math mode
    
    # Return the LaTeX formatted equation
    return equation_str.replace("+ -", "- ").strip(" +")

def plot_polynomial_with_latex(coefficients, x_range=(-10, 10)):
    # Generate x values
    x = np.linspace(x_range[0], x_range[1], 400)
    # Create the polynomial
    poly = np.poly1d(coefficients)
    # Evaluate the polynomial
    y = poly(x)
    
    # Plot the polynomial
    plt.plot(x, y, label=f'Equation: {polynomial_equation_latex(coefficients)}')
    plt.title(f'Polynomial Plot')
    plt.xlabel('x')
    plt.ylabel('y')
    
    # Display the LaTeX equation using Matplotlib's text function
    equation_latex = polynomial_equation_latex(coefficients)
    plt.text(0.45, 0.95, equation_latex, transform=plt.gca().transAxes, fontsize=12, verticalalignment='top', bbox=dict(facecolor='white', alpha=0.5))
    
    plt.grid(True)
    plt.show()

# Example usage
coefficients = [1, -3, 2]
#plot_polynomial(coefficients)
plot_polynomial_with_latex(coefficients)
```
